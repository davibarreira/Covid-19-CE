```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
pd.options.display.max_columns = 50
```

```python
df = pd.read_csv("dados/casos_coronavirus.csv",encoding='latin')
```

```python
df = df[pd.isna(df['estadoPaciente'])|(df['estadoPaciente']=='CE')]
```

```python
dcasos = df[['codigoPaciente','sexoPaciente','municipioPaciente','resultadoFinalExame','dataInicioSintomas','dataNotificacao','dataResultadoExame']]
# dcasos[dcasos['municipioPaciente']=='FORTALEZA'].pivot_table(index=['resultadoFinalExame','municipioPaciente'],aggfunc='count')
# dcasos_aux = dcasos.pivot_table(index=['resultadoFinalExame','municipioPaciente','codigoPaciente'],aggfunc='count').reset_index()

dcasos['dataCasoComputado'] = dcasos['dataInicioSintomas']
# dcasos['dataCasoComputado'].fillna(dcasos['dataNotificacao'],inplace=True)
dcasos['dataCasoComputado'].fillna(dcasos['dataResultadoExame'],inplace=True)
dcasos['ano'] = dcasos['dataCasoComputado']
dcasos['ano'] = dcasos['ano'][pd.notnull(dcasos['ano'])].apply(lambda x: int(x[0:4]))
# dcasos['dataCasoComputado'] = dcasos.loc[dcasos['ano']==202,'dataCasoComputado'].apply(lambda x: "2020"+x[4:])
dcasos.loc[dcasos['ano']==202,'dataCasoComputado'] = dcasos.loc[dcasos['ano']==202,'dataCasoComputado'].apply(lambda x: "2020"+x[4:])
dcasos.loc[dcasos['ano']==220,'dataCasoComputado'] = dcasos.loc[dcasos['ano']==220,'dataCasoComputado'].apply(lambda x: "2020"+x[4:])
dcasos.loc[dcasos['ano']==202,'ano'] = 2020
dcasos.loc[dcasos['ano']==220,'ano'] = 2020

dcasos['dataCasoComputado']= pd.to_datetime(dcasos['dataCasoComputado'])
dcasos['mes'] = pd.DatetimeIndex(dcasos['dataCasoComputado']).month
dcasos['dia'] = pd.DatetimeIndex(dcasos['dataCasoComputado']).day
dcasos['semana'] = dcasos['dataCasoComputado'].dt.week
dcasos['dia_do_ano'] = dcasos['dataCasoComputado'].dt.dayofyear
```

```python
dceara = dcasos.pivot_table(index=['resultadoFinalExame','dia_do_ano','dataCasoComputado',
                          'dia','mes','ano'],values=["codigoPaciente"],aggfunc=lambda x: len(x.unique())).reset_index()

dceara = dceara[dceara['resultadoFinalExame']=='Positivo']
dceara = dceara.sort_values(by="dia_do_ano")
dceara = dceara.sort_values(by='dataCasoComputado')
dceara.rename(columns={"codigoPaciente":"Casos"},inplace=True)
dceara["Casos_Acumulados"] = dceara.Casos.cumsum()
dceara.reset_index(drop=True,inplace=True)
dceara['Dias desde o primeiro caso'] = dceara["dia_do_ano"] - dceara.loc[0,"dia_do_ano"]
```

```python
dceara
```

```python
dceara['Cada 1 dias']   = 2**(dceara['Dias desde o primeiro caso']/1)
dceara['Cada 2 dias']   = 2**(dceara['Dias desde o primeiro caso']/2)
dceara['Cada 3 dias']   = 2**(dceara['Dias desde o primeiro caso']/3)
dceara['Cada 5 dias']   = 2**(dceara['Dias desde o primeiro caso']/5)
dceara['Cada 10 dias']  = 2**(dceara['Dias desde o primeiro caso']/10)
```

```python
from matplotlib.ticker import ScalarFormatter
sns.set(style="darkgrid")

fig, ax = plt.subplots(figsize=(15,10))
# ax.set(yscale="log")
# sns.lineplot(data=dceara,x='Dias desde o primeiro caso',y='Casos_Acumulados',ax=ax,err_style=None)
# sns.lineplot(data=dceara,x='Dias desde o primeiro caso',y='Cada 1 dias',ax=ax,err_style=None,)
plt.plot(dceara['Dias desde o primeiro caso'],dceara['Casos_Acumulados'])
plt.plot(dceara['Dias desde o primeiro caso'],dceara['Cada 1 dias'],'--',color='lightgray')
plt.plot(dceara['Dias desde o primeiro caso'],dceara['Cada 2 dias'],'--',color='lightgray')
plt.plot(dceara['Dias desde o primeiro caso'],dceara['Cada 3 dias'],'--',color='lightgray')
plt.plot(dceara['Dias desde o primeiro caso'],dceara['Cada 5 dias'],'--',color='lightgray')
plt.plot(dceara['Dias desde o primeiro caso'],dceara['Cada 10 dias'],'--',color='lightgray')
plt.yscale("log")
plt.ylim(top=10**5)
# plt.yticks(np.range(6))
plt.text(12,8000, 'cada 1 dias', fontsize=16,
               rotation=70, rotation_mode='anchor',color="lightgray")
plt.text(25,10000, 'cada 2 dias', fontsize=16,
               rotation=50, rotation_mode='anchor',color="lightgray")
plt.text(35,4500, 'cada 3 dias', fontsize=16,
               rotation=40, rotation_mode='anchor',color="lightgray")
plt.text(40,300, 'cada 5 dias', fontsize=16,
               rotation=20, rotation_mode='anchor',color="lightgray")
plt.text(40,20, 'cada 10 dias', fontsize=16,
               rotation=10, rotation_mode='anchor',color="lightgray")
ax.yaxis.set_major_formatter(ScalarFormatter())
```
